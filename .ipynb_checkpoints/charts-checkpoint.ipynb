```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nycWaterConsumption1979-2020.csv')
```

```python
df.head()
```

```python
year = df["year"]
```

```python
population = df["population"]/1000000
```

```python
consumption_million_gallons_per_day = df["consumption_million_gallons_per_day"]
```

```python
consumption_gallons_per_capita_per_day = df["consumption_gallons_per_capita_per_day"]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(year, 
             population, 
             color= "purple",
             label="Population")

axes[1].plot(year,
             consumption_gallons_per_capita_per_day,
             color= "dodgerblue",
             label="Water Consumption");


axes[0].set_xlabel("Year",
                   fontsize=12)

axes[0].set_ylabel("Population in Millions",
                   fontsize=12)

axes[0].ticklabel_format(style='plain')


axes[1].set_xlabel("Year",
                   fontsize=12)

axes[1].set_ylabel("Gallons Per Capita Per Day",
                   fontsize=12)

fig.suptitle("NYC Population and Water Consumption 1979-2020",
             fontsize=20,
             fontweight="bold")

fig.legend(ncol=2,
           bbox_to_anchor=(0.5,0.01),
           loc="upper center")

plt.annotate('Data source: City of New York',
            xy = (-1.5, -0.24),
            xycoords='axes fraction',
            fontsize=9)

plt.annotate('@albertoscode',
            xy = (.75, -0.24),
            xycoords='axes fraction',
            fontsize=9)

plt.subplots_adjust(left=0.09,
                    right=0.89,
                    bottom=0.11, 
                    top=0.89, 
                    wspace=0.37, 
                    hspace=0.37)

plt.savefig("chart1.jpg",
            dpi=150,
            bbox_inches='tight')

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 4))

ax2 = ax1.twinx()

ax1.plot(year,
         population,
         color= "purple",
         label="Population")

ax2.plot(year,
         consumption_gallons_per_capita_per_day,
         color= "dodgerblue",
         label="Water Consumption");

ax1.set_xlabel("Year",
               fontsize=12)

ax1.set_ylabel("Population in Millions",
               fontsize=12)

ax1.ticklabel_format(style='plain')
               

ax2.set_ylabel("Gallons Per Capita Per Day",
               fontsize=12)
               
fig.suptitle("NYC Population and Water Consumption 1979-2020",
             fontsize=20,
             fontweight="bold")

fig.legend(ncol=2,
           bbox_to_anchor=(0.5,0.01),
           loc="upper center")

plt.annotate('Data source: City of New York',
            xy = (-0.06, -0.35),
            xycoords='axes fraction',
            fontsize=9)

plt.annotate('@albertoscode',
            xy = (.90, -0.35),
            xycoords='axes fraction',
            fontsize=9)

plt.savefig("chart2.jpg",
            dpi=150,
            bbox_inches='tight')

plt.show()
```
