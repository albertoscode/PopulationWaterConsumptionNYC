```python
df_new_column_names = df.set_axis(["year", 
                                   "population",
                                   "consumption_million_gallons_per_day",
                                   "consumption_gallons_per_capita_per_day"], 
                                    axis=1)
```

```python
df_new_column_names.head(50)
```

```python
df_new_column_names.info()
```

```python
df_roundup_population = df_new_column_names.copy()
```

```python
df_roundup_population.head()
```

```python
df_roundup_population['population'] = df_roundup_population['population'].apply(np.ceil).astype(int)
```

```python
df_roundup_population.head(25)
```

```python
df_roundup_cmg_pd = df_roundup_population.copy()
```

```python
df_roundup_cmg_pd.head(50)
```

```python
df_roundup_cmg_pd['consumption_million_gallons_per_day'] = df_roundup_cmg_pd['consumption_million_gallons_per_day'].apply(np.ceil).astype(int)
```

```python
df_roundup_cmg_pd.head(25)
```

```python
df_nyc_water_consumption_1979_2020 = df_roundup_cmg_pd
```

```python
df_nyc_water_consumption_1979_2020.head(50)
```

```python
df_nyc_water_consumption_1979_2020.info()
```

```python
df_nyc_water_consumption_1979_2020['year'] = df_nyc_water_consumption_1979_2020['year'].astype('string')
```

```python
df_nyc_water_consumption_1979_2020.info()
```

```python
df_nyc_water_consumption_1979_2020["consumption_million_gallons_per_day"].describe()
```

```python
df_nyc_water_consumption_1979_2020["consumption_gallons_per_capita_per_day"].describe()
```
